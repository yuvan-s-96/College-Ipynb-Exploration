```python
import requests

from bs4 import BeautifulSoup
import pandas as pd
import time

# URL of the website to scrape
url = "https://www.imdb.com/chart/top"

# Send an HTTP GET request to the website
response = requests.get(url)

# Parse the HTML code using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the relevant information from the HTML code
movies = []
for row in soup.select('tbody.lister-list tr'):
    title = row.find('td', class_='titleColumn').find('a').get_text()
    year = row.find('td', class_='titleColumn').find('span', class_='secondaryInfo').get_text()[1:-1]
    rating = row.find('td', class_='ratingColumn imdbRating').find('strong').get_text()
    movies.append([title, year, rating])
print(movies)
# Store the information in a pandas dataframe
df = pd.DataFrame(movies, columns=['Title', 'Year', 'Rating'])

# Add a delay between requests to avoid overwhelming the website with requests
time.sleep(1)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the list of countries
table = soup.find('table', {'class': 'wikitable'})

# Initialize a list to hold country data
countries = []

# Iterate through each row in the table
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) > 1:  # Ensure there are enough columns
        # Extract the country name and population
        country = cols[1].get_text(strip=True)
        population = cols[2].get_text(strip=True)
        countries.append((country, population))

# Print out the list of countries and their populations
for country, population in countries:
    print(f"Country: {country}, Population: {population}")

```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Climb The Ladder page with web scraping interview questions
url = "https://climbtheladder.com/web-scraping-interview-questions/"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all question blocks (adjust this based on the actual HTML structure)
    question_blocks = soup.find_all('div', class_='entry-content')  # Use the correct class for the content block

    for block in question_blocks:
        # Extract questions and answers (update the tags and classes as necessary)
        questions = block.find_all('h3')  # Assuming questions are in <h3> tags
        answers = block.find_all('p')     # Assuming answers are in <p> tags

        for question, answer in zip(questions, answers):
            print(f"Question: {question.get_text(strip=True)}")
            print(f"Answer: {answer.get_text(strip=True)}")
            print("=" * 50)
else:
    print("Failed to retrieve the webpage.")
```
