MISSING DATA
Like c program and java the missing value is null,similarly null values in pandas are NaN(Not a Number).
NaN is a special floating point represented as null value

```python
df=pd.DataFrame({'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]})
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.fillna(value="Fill")
```

```python
df['A'].fillna(value=df['A'].mean())
```
