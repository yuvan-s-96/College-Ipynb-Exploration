```python
from transformers import T5Tokenizer, AutoModelForSeq2SeqLM

# Load tokenizer using T5Tokenizer instead of AutoTokenizer
tokenizer = T5Tokenizer.from_pretrained("flax-community/t5-recipe-generation")
model = AutoModelForSeq2SeqLM.from_pretrained("flax-community/t5-recipe-generation")

# Save the tokenizer and model locally
save_directory = "./local_model_directory"
tokenizer.save_pretrained(save_directory)
model.save_pretrained(save_directory)
```
